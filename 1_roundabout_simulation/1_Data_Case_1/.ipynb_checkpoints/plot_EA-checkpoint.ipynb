## Vehicle Parameters and Functions for Computing the Vertices of Vehicle

```python
l_veh_ev = 4.3 # vehicle length, m
w_veh_ev = 1.8 # vehicle width, m
l_veh_sv = 4.3
w_veh_sv = 1.8
target_x = 50

def rota_rect(box, theta, x, y):
    """
    :param box: vertices
    :param theta: angle
    :param x: x center
    :param y: y center
    
    """
    box_matrix = np.array(box) - np.repeat(np.array([[x, y]]), len(box), 0)
    theta = -theta
    rota_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                            [np.sin(theta), np.cos(theta)]])

    new = box_matrix.dot(rota_matrix) + np.repeat(np.array([[x, y]]), len(box), 0)
    
    x = [new[0][0], new[1][0], new[2][0], new[3][0], new[0][0]]
    y = [new[0][1], new[1][1], new[2][1], new[3][1], new[0][1]]
    return x, y
```

## Compute the Minimal Distance Between EV and SVs by different planners

```python
Min_Dis_SV = list( )
for k in range(len(State_SV[0])):
    x = State_SV[0, k]
    y = State_SV[1, k]
    box = [(x - l_veh_sv/2, y + w_veh_sv/2), (x + l_veh_sv/2, y + w_veh_sv/2), (x + l_veh_sv/2, y - w_veh_sv/2), (x - l_veh_sv/2, y - w_veh_sv/2)]
    heading = State_SV[2, k]
    x_SV, y_SV = rota_rect(box, heading, x, y)
   
    x = State_EV[0, k]
    y = State_EV[1, k]
    box = [(x - l_veh_ev/2, y + w_veh_ev/2), (x + l_veh_ev/2, y + w_veh_ev/2), (x + l_veh_ev/2, y - w_veh_ev/2), (x - l_veh_ev/2, y - w_veh_ev/2)]
    heading = State_EV[2, k]
    x_EV, y_EV = rota_rect(box, heading, x, y)
    
    pSV = Polygon([(x_SV[0],y_SV[0]), (x_SV[1],y_SV[1]), (x_SV[2],y_SV[2]), (x_SV[3],y_SV[3])])
    
    pEV   = Polygon([(x_EV[0], y_EV[0]), (x_EV[1], y_EV[1]), (x_EV[2], y_EV[2]), (x_EV[3], y_EV[3])])

    Min_Dis_SV.append(pEV.distance(pSV))
```

```python
dis_x = abs(State_EV[0, :] - target_x)
index_time = np.argmin(dis_x)
time_fin = t[index_time]
```
